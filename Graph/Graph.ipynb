```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from IPython.display import display, HTML

x = [1, 2, 3, 4, 5]
y = [6, 7, 8, 9, 10]


plt.title('Scatter plot of x and y')
plt.scatter(x, y)
plt.show()

plt.title('Showing the point in a line graph')
plt.plot(x, y)
plt.show()


plt.title('Showing the point in a line graph with scatter plot')
plt.scatter(x,y)
plt.plot(x, y)
plt.show()


x2= [11, 12, 13, 14, 15]
y2= [16, 17, 18, 19, 20]
plt.title('Showing two different graphs')
plt.scatter(x2,y2)
plt.plot(x2,y2)
plt.scatter(x,y)
plt.plot(x,y)
plt.show()

display(HTML("<h1>' color mark '</h1>"))
plt.title("Showing graphs with colors and marks")
plt.plot(x,y,"ro")
plt.plot(x2,y2,"g--")
plt.show()


display(HTML("<h2>' Functional Graph '</h2>"))
plt.title("Y = X^3")
ara = np.array([1,2,3,4])
x4 = np.arange(0 , 5 , 0.1)
y4 = x4 **3
plt.plot(x4,y4,"bo")
plt.show()



display(HTML("""
<article>
    <section>
        <h2>Using Color and Marker in Matplotlib</h2>
        <p>To customize plots in Matplotlib, various options can be utilized:</p>
        <ul>
            <li><strong>Linewidth:</strong> Adjust the line width using <code>plt.plot(..., linewidth=N)</code>.</li>
            <li><strong>Axis Labels:</strong> Label the X and Y axes with <code>plt.xlabel('X-axis')</code> and <code>plt.ylabel('Y-axis')</code>.</li>
            <li><strong>Grid:</strong> Add a grid to the plot with <code>plt.grid()</code>.</li>
            <li><strong>Axis Range:</strong> Define the range of the axes using <code>plt.axis([x1, x2, y1, y2])</code>.</li>
            <li><strong>Color:</strong> Adding text to the plot using <code>plt.text(2, 80, "Text")</code>.</li>
        </ul>
    </section>
    <section>
        <h2>Understanding Legends</h2>
        <p>Legends provide a meaningful explanation of the various elements that appear in the chart, helping to differentiate between the datasets or categories represented.</p>
    </section>
    <section>
        <h2>Marker Notations</h2>
        <p>Below is a list of markers available in Matplotlib's plot function:</p>
        <ol>
            <li>'o' = Circle</li>
            <li>'s' = Square</li>
            <li>'p' = Pentagon</li>
            <li>'h' = Hexagon</li>
            <li>'x' = X-mark</li>
            <li>'+' = Plus</li>
            <li>'D' = Diamond</li>
            <li>'^' = Triangle Up</li>
            <li>'v' = Triangle Down</li>
            <li>'>' = Triangle Right</li>
            <li>'<' = Triangle Left</li>
            <li>'*' = Star</li>
            <li>'1' = Tri Down</li>
            <li>'2' = Tri Up</li>
            <li>'3' = Tri Left</li>
            <li>'4' = Tri Right</li>
            <li>'8' = Octagon</li>
            <li>'P' = Plus (filled)</li>
            <li>'H' = Hexagon1</li>
            <li>'X' = X (filled)</li>
            <li>'|' = Vline</li>
            <li>'_' = Hline</li>
        </ol>
    </section>
</article>
"""))

plt.title("Previous Function using color and marker")
plt.xlabel("X")
plt.ylabel("X^3")
plt.text(8, 80, "DBP")
ara = np.array([1,2,3,4]) 
x5 = np.arange(0 , 5 , 0.1)
y5 = x4 **3
y6 = x5 **2
plt.grid()
plt.axis([0 , 100 , 0 , 100])


plt.plot(x5,y5, color = "red" , marker = "p" , linewidth = 2.5 , label = "X^3")
plt.plot(x5, y6, color = "blue" , marker = "s" , linewidth = 2.5 , label = "X^2")
plt.legend() #DBP_NOTES To differentiate between the two graphs
plt.show()


display(HTML("""
<article>
    <section>
        <h2>Creating a Bubble Chart with Matplotlib</h2>
        <p>Bubble charts are a type of scatter plot that allows the visualization of three dimensions of data. Each point's x and y position corresponds to two of the dimensions, and the size of the marker (the 'bubble') represents the third dimension.</p>
        <p>In Matplotlib, you can create a bubble chart using the <code>plt.scatter()</code> function. Here's how to customize your bubble chart:</p>
        <ul>
            <li><strong>X and Y Coordinates:</strong> These are the basic inputs for the scatter plot, determining the position of each bubble.</li>
            <li><strong>Size (s):</strong> The size of each bubble can be specified with the <code>s</code> parameter. This can be a single value or an array of sizes, one for each point.</li>
            <li><strong>Color (c):</strong> You can specify the color of the bubbles using the <code>c</code> parameter. This can also be a single color or an array to color each bubble differently.</li>
            <li><strong>Alpha:</strong> Transparency of the Bubbles can be adjusted using the <code>alpha</code> parameter, which takes a value between 0 (completely transparent) and 1 (completely opaque).</li>
            <li><strong>edgecolor:</strong> The color of the bubble's edge can be specified using the <code>edgecolor</code> parameter.</li>
        </ul>
        <p>Here's an example of creating a bubble chart that visualizes the relationship between the year, average salary, and population size:</p>
        <pre>
plt.title("Bubble-Chart")
year = [2012, 2013, 2014, 2015, 2016, 2017]
salary = [12, 13, 14, 17, 19, 20]
population = [100, 120, 180, 250, 300, 370]

plt.scatter(year, salary, s=population)
plt.show()
        </pre>
        <p>This chart uses the <code>year</code> and <code>salary</code> as the x and y coordinates, respectively, and the <code>population</code> size to determine the size of each bubble.</p>
    </section>
</article>
"""))

plt.title("Bubble-Chart")
year = [2012, 2013, 2014, 2015, 2016, 2017]
salary = [12, 13, 14, 17, 19, 20]
population = [100, 120, 180, 250, 300, 370]

plt.scatter(year , salary , s=population , c = 'r' , alpha=0.5)
plt.title("Bubble-Chart")
plt.show()

c=np.arange(len(year))
c[1]=c[0]
plt.scatter(year , salary , s=population , c = c , alpha=0.5 , edgecolor='black')
for i in range(len(year)):
    #DBP_HIGHLIGHTES here + 0.1 is used to avoid overlapping of the text with the bubble
    plt.text(year[i] + 0.12 , salary[i] + 0.12 , population[i])

plt.show()



display(HTML("""
<article>
    <section>
        <h2>Pie Chart Overview</h2>
        <p>A pie chart is a circular statistical graphic, which is divided into slices to illustrate numerical proportion. In this example, we create a pie chart to represent programming language preferences.</p>
    </section>
    <section>
        <h3>plt.pie(</h3>
        <ul>
            <li><strong>sizes:</strong> Represents the proportion of each slice. In this case, [25, 35, 20, 20].</li>
            <li><strong>colors:</strong> Specifies the color of each slice. Used colors are ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue'].</li>
            <li><strong>labels:</strong> Describes what each slice represents. Labels used are ['Python', 'C++', 'Ruby', 'Java'].</li>
            <li>autopct='%.nf% %': %.n% = Displays the percentage to n decimal places , then % adds %sign </li>
            <li>explode: To highlight a particular slice , explode[0.1,0,0,0] means highlighting the first slice
            explode range is from 0 to 1</li>
            )
        </ul>
    </section>
</article>
"""))

sizes = [25, 35, 20, 20]
colors = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue']
labels = ['Python', 'C++', 'Ruby', 'Java']
plt.title("Pie Chart")
plt.axis("equal")
plt.pie(sizes , colors = colors , labels=labels , autopct="%.2f%%" , explode=[0.1,0,0.2,0])
plt.show()
```
