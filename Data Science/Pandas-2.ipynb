````python
import pandas as pd
from IPython.display import display, HTML
iris =pd.read_csv("https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data")
df = iris.copy()


display(HTML("<strong>\n Column Operations</strong>"))
df.columns = ['sl', 'sw', 'pl', 'pw', 'flower_type']
print("Using df.drop(0)")
df.drop(0)

print("#Here the original data is not changed , a copy of the data is changed with the drop function")
print(df.head() )

print("\nTo show the change , storing the drop function in a variable")
change = df.drop(0)
print(change.head())

print("\n\nTo show change in original data , we have to use inplace = True")
df.drop(0, inplace = True)
print(df.head())

print("\nTo show the all the labels")
df.index

print(df.index[0])


print("\nTo drop the first two rows[label 0 and 1]")
df.drop(df.index[ [0,1] ], inplace = True)
df.head()


print("\n Logical Operation on column")
print(df.sl > 5)

print("\n Rows of Column sl > 5")
print(df[df.sl > 5])

print("\n Rows of flower_type == 'Iris-setosa'")
print(df[df.flower_type == 'Iris-setosa'])


print("\n Description of the Rows of flower_type == 'Iris-setosa \n")
df[df.flower_type == 'Iris-setosa'].describe()

print("\n Head \n" , df.head())
print("\n iloc[0] \n" , df.iloc[0])
#print("\n loc[0]" , df.loc[0]) #DBP_NOTES error as there is no label 0

df.loc[0] = [1 ,2 ,3 ,4 , 'Iris-setosa']
print(df.tail())


print("\n Resetting Index[var.reset_index(),\n")
df.reset_index(drop = True, inplace = True)
df.head()
display(HTML("<strong>\n Deleting a column , But not original</strong>")) #DBP_HIGHLIGHTES well by using df.drop('sl',  axis = 1) , it will not delete the original column
df.drop('sl',  axis = 1)
df.head()


display(HTML("<strong>\n Deleting a column but original</strong>"))
df.drop('sl' , axis=1 , inplace = True)
df.head()
df.describe()


display(HTML("<strong>\n Standard Way: del </strong> var_panda['col_name]"))
del df['sw']
df.head()
df.describe()


display(HTML("<strong>\n Adding a column</strong>"))
display(HTML("<strong>\n Syntax: var_panda['new_col_name'] = value</strong>"))
display(HTML("<strong>\n A column that gives the difference of pl and pw</strong>"))
df["diff_pl_pw"] = df["pl"] - df["pw"]
df.head()


````

````python
import numpy as np
import pandas as pd
from IPython.display import display, HTML
iris =pd.read_csv("https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data")
df = iris.copy()
df.columns = ['sl', 'sw', 'pl', 'pw', 'flower_type']

display(HTML("<strong style='color:white;background:black'>\n Handling Column with NaN[Not a Number] Entries</strong>"))

display(HTML("<strong style='background:yellow:'>\n Creating Nan Entries ; np.nan</strong>"))
df.iloc[2:4 , 1:3] = np.nan
print(df.head())

df.describe()

display(HTML("<strong>\n Method1: Dropping the nan : df.dropna(inplace=True)</strong>"))
dropna=df.dropna()
print(dropna) 

display(HTML("<strong style='background:black;color:white'>\n Method2: Filling the nan with the mean of corresponding column :\n df.colname.fillna(df. colname.mean())</strong>"))
df.sw.fillna(df.sw.mean() , inplace=True)
df.pl.fillna(df.pl.mean() , inplace=True)
df.head()


display(HTML("<strong style='background:black;color:white'>\n Handling Strings in Columns(Y/N)</strong>"))
df["Gender"] = "Female" #DBP_NOTES Adding a new Column
df.iloc[0:5 , 5] = "Male" #DBP_NOTES Changing the values of first 5 rows of 6th columns to Male
print(df.head())
print(df.tail())

display(HTML("<strong style='background:black;color:white'>\n Changing the string into numeric 0/1 using a function</strong>"))
def f(s) :
    if s == "Male":
        return 0
    else:
        return 1

df["Sex"] = df.Gender.apply(f)
print(df.head())
print(df.tail())

display(HTML("<strong style='background:black;color:white'>\n After changing to Numeric , deleting the Gender Column to remover redaundency , as it is converted to Sex Column</strong>"))
del df["Gender"]
print(df.head())
````

# Time Execution (Numpy Faster than Array)

1. t1 means start time and t2 means end time
2. <i>function defition format is <p style="color:yellow">def function_name():</p> </i>
3.  Difference between rang() and np.arange() is that range() is used in list and np.arange() is used in numpy array

4. Storing function values in variable
  <ul style="color:yellow">
  <li>t_list = addition_using_list()</li>
  <li>t_numpy = addition_using_numpy() </li>
  </ul>

````python
import time #DBP_HIGHLIGHTES time is used to get the time of execution of code
import numpy as np

size = 1000000

def addition_using_list():
    t1 = time.time() #time.time() means current time.
    a = range(size)
    b = range(size)
    c = [ a[i] + b[i] for i in range(size)]
    t2 = time.time()
    return t2 - t1

    #DBP_NOTES here t1 means start time and t2 means end time

def addition_using_numpy():
    t1 = time.time()
    a = np.arange(size)
    b = np.arange(size)
    c = a + b
    t2 = time.time()
    return t2 - t1

t_list = addition_using_list()
t_numpy = addition_using_numpy()

print("List Time: ", t_list * 1000)
print("Numpy Time: ", t_numpy * 1000)




````
