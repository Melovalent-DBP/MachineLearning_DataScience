# Time Execution (Numpy Faster than Array)

1. t1 means start time and t2 means end time
2. <i>function defition format is <p style="color:yellow">def function_name():</p> </i>
3.  Difference between rang() and np.arange() is that range() is used in list and np.arange() is used in numpy array

4. Storing function values in variable
  <ul style="color:yellow">
  <li>t_list = addition_using_list()</li>
  <li>t_numpy = addition_using_numpy() </li>
  </ul>

````python
import time #DBP_HIGHLIGHTES time is used to get the time of execution of code
import numpy as np

size = 1000000

def addition_using_list(): 
    t1 = time.time() #time.time() means current time.
    a = range(size) 
    b = range(size)
    c = [ a[i] + b[i] for i in range(size)]
    t2 = time.time() 
    return t2 - t1

    #DBP_NOTES here t1 means start time and t2 means end time

def addition_using_numpy():
    t1 = time.time() 
    a = np.arange(size) 
    b = np.arange(size)
    c = a + b
    t2 = time.time()
    return t2 - t1

t_list = addition_using_list()
t_numpy = addition_using_numpy()   

print("List Time: ", t_list * 1000)
print("Numpy Time: ", t_numpy * 1000)




````

# Broadcasting in Numpy
## <b style = "color:black; background:orange">It is used to perform operations on arrays of different shapes.</b>

- ## Compatibility
### 1. <b style = "color:black; background:white"> x[3,2] , y[3,2] </b>&nbsp; Perfectly Compatibile . So , + - * / ** can be possible.
### One of them is 1 <br>

### 1. <b style = "color:black; background:white">x[<i>3</i>,3] , y[<i>3</i>,2] </b>&nbsp; Not Compatibile
- ### <b style = "color:black; background:LightGreen">Solution is np.transpose(y)</b>
- ### <b style = "color:black; background:LightGreen"> np.reshape(y , (2,3) , order = 'C' / 'F' / 'A') </b>
#### <b style = "color:black; background:orange"> Careful dimension multplication , [For Instance: if arr(m,n) , reshape will be (n,m)] should be same </b> 
&nbsp; 
### 2. <b style = "color:black; background:white">x[<i>3</i>,3] , y[<i>3</i>] / y[1,<i>3</i>] </b> &nbsp; Compatible . So , + - * / ** can be possible.</b>

- ## Real Time Broadcasting
1. ### <b style = "color:black; background:LightGreen"> Before Broadcasting:


````python
import numpy as np


print("Perfectly Compatible","\n")
x= np.random.randint(1 , 10 , (3,3))
y = np.random.randint(1 , 10 , (3,3))

print(x,"\n")
print(y , "\n" )
print(x ** y)

print('\n\n')
print("Partially Compatible","\n")
x1= np.random.randint(1 , 10 , (3,3))
y1 = np.random.randint(1 , 10 , (3))

print(x1 , "\n")
print(y1 , "\n")
print(x1 * y1 , "\n\n" , x - y)
print("\n")

print("Not Compatible","\n")
x2 = np.random.randint(1 , 10 , (3 , 2))
y2 = np.random.randint(1 , 10 , (2 , 3))

print("\n" , x2 , "\n")
print("\n" , y2 , "\n")



#DBP_NOTES as y2 and x2 has different dimension , so the solution is 
#DBP_HIGHLIGHTES TRANSPOSE of y2
print("Converting y2 in its transpose MAtrix (3 * 2)" , "\n")
y3 = np.transpose(y2)
print(y3,"\n")
print("Now the operations can be possible" , "\n")
print(x2 * y3)

print("\n")

#DBP_HIGHLIGHTES Reshaping can be possible when the dimension multiplication is same
arr_reshaped = np.reshape(x2 , (2,3))
arr_reshaped

````
